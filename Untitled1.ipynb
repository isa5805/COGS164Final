```python
#Make Dummy Columns For Categorical Variables 

def yes_int_converter(x):
    if x=='No':
        return 0
    elif x=='Yes':
        return 1

df['Question_1'] = df['Question_1'].apply(yes_int_converter)
df['Question_2'] = df['Question_2'].apply(yes_int_converter)
df['Question_15'] = df['Question_15'].apply(yes_int_converter)
df['Question_4'] = df['Question_4'].apply(yes_int_converter)
df['Question_26'] = df['Question_26'].apply(yes_int_converter)
```

```python
def gender_converter(x):
    if x=='Prefer not to say':
        return 'Prefer_not_to_say'
    elif x == 'Non-binary':
        return 'Non_binary'
    else:
        return x
df['Gender'] = df['Gender'].apply(gender_converter)
```
