After dropping the first choice of columns, the next thing I looked at was to see if all Age responses were recorded as numbers. They were not so I created a function to get rid of non numeric  characters.  

```python
#getting rid of string from responses of Age
def standardize_age(string):
    
    string = string.lower()
    string = string.strip()
    
    if 'feeling' in string:
        output = '22'
   

    else:
        output = string
    
    return output
df['Age'] = df['Age'].apply(standardize_age)

df['Age'] = pd.to_numeric(df['Age'])
```

```python
#responses varied, grouped correctly
def standardize_major(string):
    
    string = string.lower()
    string = string.strip()
    
    if 'cog' in string:
        output = 'Cognitive Science'
    elif 'computer' in string:
        output = 'Computer Science'
    elif 'cs' in string:
        output = 'Computer Science'
    elif 'software' in string:
        output = 'Computer Science'
    elif 'math' in string:
        output = 'Math'
    elif 'electrical' in string:
        output = 'Elecrical Engeneering'
    elif 'bio' in string:
        output = 'Biology'
    elif 'uman' in string:
        output = 'Human Biology'
    elif 'iochem' in string:
        output = 'Biochemisry'
    elif 'data' in string:
        output = 'Data Science'
    elif 'poli' in string:
        output = 'Political Science'
    # Otherwise, if uncaught - keep as is
    elif 'sychology' in string:
        output = 'Psychology'
    elif 'nglish' in string:
        output = 'English'
    elif 'sociology' in string:
        output = 'Sociology'
    elif 'communication' in string:
        output = 'Communication'
    else:
        output = string
    
    return output
```

Here I begin to quantify the answers. I turn yes and no responses to 0s and 1s. I modify responses again to help my code run smoother. Lastly, I perform one hot encoding to create dummy columns with categorical responses that have no numeric value.

For instance for question 10, if I answered that vacations make me plan less, I would get a 1 in this new vacation dummy column but 0s for the other response dummy columns such as other obligations or mental health. 

```python
def standardize_q15(string):
    
    string = string.lower()
    string = string.strip()
    
    if 'yes' in string:
        output = 'Yes'
    elif 'no' in string:
        output = 'No'
   
    else:
        output = string
    
    return output
df['Question_15'] = df['Question_15'].apply(standardize_q15)
df['Question_15'].unique()
```

```python
#Make Dummy Columns For Categorical Variables 

def yes_int_converter(x):
    if x=='No':
        return 0
    elif x=='Yes':
        return 1

df['Question_1'] = df['Question_1'].apply(yes_int_converter)
df['Question_2'] = df['Question_2'].apply(yes_int_converter)
df['Question_15'] = df['Question_15'].apply(yes_int_converter)
df['Question_4'] = df['Question_4'].apply(yes_int_converter)
df['Question_26'] = df['Question_26'].apply(yes_int_converter)
```

```python
def gender_converter(x):
    if x=='Prefer not to say':
        return 'Prefer_not_to_say'
    elif x == 'Non-binary':
        return 'Non_binary'
    else:
        return x
df['Gender'] = df['Gender'].apply(gender_converter)
```

Since beginning the dive in the data, I realized that some questions overlap, especially the Intolerance of Uncertainty questions. 

Because we are trying to find out if the use of a planner correlates with how comfortable someone is with uncertainty, because uncertainty here is the response variable, including questions that highly correlate with one another would mess up the model. So I chose one question out of the eleven and tried to see if the responses of that question correlated with the responses of the remaining questions.

```python
#one of the anxiety questions at a time
import patsy
import statsmodels.api as sm
import scipy.stats as stats
from scipy.stats import ttest_ind, chisquare, normaltest

temp = df.drop(['Major','Question_16', 'Question_18', 'Question_19', 'Question_20', 'Question_21',
               'Question_22', 'Question_23', 'Question_24', 'Question_25', 'Question_28'], axis = 1)

temp['Question_3'] = df['Question_3'].apply(standardize_q3)
temp = pd.get_dummies(temp, columns=['Question_3'])
```
